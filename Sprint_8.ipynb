Importamos las librerias necesarias

```python
import pandas as pd
import re
from collections import defaultdict

```

Definimos una función auxiliar para reparar los errores de codificación.

```python
def reparar_codificacion(texto) -> str:
    try:
        return texto.encode("cp1252").decode("utf-8")
    except (UnicodeEncodeError, UnicodeDecodeError):
        return texto
```

Leemos el fichero excel y lo cargamos en un DataFrame de pandas. Ignoramos las 2 primeras filas y usamos la tercera fila como cabecera.
Además, establecemos la primera columna como índice del DataFrame.
Después hacemos una copia del DataFrame original para trabajar sobre ella y no modificar el original.

```python
df = pd.read_excel("sprint8_complex.xlsx", header=3, index_col=0)
nuevo_df = df.copy()
```

Comprobamos y reparamos la codificación del texto de las cabeceras de las columnas.

```python
cabeceras_originales = nuevo_df.columns
print(f"Cabeceras originales: {cabeceras_originales.tolist()}")
cabeceras_reparadas = cabeceras_originales.map(
    lambda x: reparar_codificacion(x) if isinstance(x, str) else x
)

cabeceras_modificadas = cabeceras_originales[cabeceras_originales != cabeceras_reparadas]

if len(cabeceras_modificadas) > 0:
    print(
        f"Se han reparado las cabeceras de las columnas "
        f"{cabeceras_modificadas.tolist()} debido a problemas de codificación."
    )
    nuevo_df.columns = cabeceras_reparadas
```

Reparamos el contenido de las columnas de texto.

```python
for columna in nuevo_df.select_dtypes(include=["object", "string"]):
    serie_original = nuevo_df[columna]

    serie_reparada = serie_original.map(
        lambda x: reparar_codificacion(x) if isinstance(x, str) else x
    )

    reparados = (serie_original != serie_reparada).sum()

    if reparados > 0:
        print(
            f"Se han reparado {reparados} valores en la columna "
            f"'{columna}' debido a problemas de codificación."
        )
        nuevo_df[columna] = serie_reparada
```

Obtenemos las filas duplicadas exactas y las eliminamos.

```python
filas_duplicadas = nuevo_df[nuevo_df.duplicated(keep="first")]

if not filas_duplicadas.empty:
    print("Detectadas filas duplicadas exactas:")
    print(filas_duplicadas)
else:
    print("No se han encontrado filas duplicadas exactas.")

filas_iniciales = len(nuevo_df)
nuevo_df.drop_duplicates(inplace=True)

print(f"\nSe han eliminado {filas_iniciales - len(nuevo_df)} filas duplicadas exactas.")
```

Buscamos todos aquellos DNIs que aparecen más de una vez y eliminamos todas las filas en las que aparecen, ya que no es posible determinar cual de las filas (o ninguna) tiene los datos correctos.

```python
duplicados = nuevo_df[nuevo_df.duplicated(subset=["DNI"], keep=False)]

if duplicados.empty:
    print("No se han encontrado DNIs duplicados.")
else:
    for dni, grupo in duplicados.groupby("DNI"):
        print(f"\nFilas con el DNI duplicado {dni}:")
        print(grupo)

        nuevo_df.drop(index=grupo.index, inplace=True)
        print(f"Se han eliminado {len(grupo)} filas con el DNI {dni}.")
```

Buscamos y eliminamos todas las filas que no tienen la informacion minima necesaria (Nombre, Apellidos y DNI) rellenada y sin simbolos.

```python
mask_na = (
    df["Nom"].isna()
    & df["Cognoms"].isna()
    & df["DNI"].isna()
)

if mask_na.any():
    print("Las siguientes filas contienen valores NA en los campos Nom, Cognoms y DNI y se eliminarán:\n")
    print(df.loc[mask_na])

df = df.loc[~mask_na].copy()

print(f"\nSe han eliminado {mask_na.sum()} filas.\n")

mask_simbolos = (
    df["Nom"].fillna("").str.fullmatch(r"[^A-Za-zÀ-ÿ0-9]+")
    | df["Cognoms"].fillna("").str.fullmatch(r"[^A-Za-zÀ-ÿ0-9]+")
    | df["DNI"].fillna("").str.fullmatch(r"[^A-Za-zÀ-ÿ0-9]+")
)

if mask_simbolos.any():
    print("Las siguientes filas contienen símbolos no válidos en alguno de los campos Nom, Cognoms o DNI y se eliminarán:\n")
    print(df.loc[mask_simbolos])

df = df.loc[~mask_simbolos].copy()

print(f"\nSe han eliminado {mask_simbolos.sum()} filas.\n")
```

```python
Comprobamos el contenido de las columas para detectar y corregir valores imposibles.
Para ello, definimos los rangos de valores permitidos para cada columna y los tipos de datos esperados.
```

```python
ENTEROS = {
    "Dia de Naixement": (1, 31),
    "Mes de Naixement": (1, 12),
    "Any de Naixement": (1900, 2026),
    "Salari mensual": (0, None),
    "Nombre_fills": (0, None),
    "Km_anuals": (0, None),
}

FLOATS = {
    "Consum_mitja_L_100km": (0, None),
    "Temperatura_mitjana_ciutat": (-273.15, None),
}

STRINGS = {
    "Gènere": {"H", "D"},
    "Grup Professional": {
        "Grup A",
        "Grup B",
        "Grup C",
        "Grup D",
    },
}

BOOLEANOS = [
    "Te_cotxe",
    "Fills",
    "No Fills",
]
```

Creamos un diccionario para almacenar los errores encontrados en cada fila del DataFrame.
Validamos los valores de las columnas según los rangos y tipos definidos. Empezamos por los enteros.

```python
errores = defaultdict(list)

for columna, (minimo, maximo) in ENTEROS.items():

    valores = pd.to_numeric(df[columna], errors="coerce")

    # No es un número entero
    mask = valores.isna()

    for indice in df.index[mask]:
        errores[indice].append(
            f"{columna}: '{df.at[indice, columna]}' no es un número entero."
        )

    # Menor que el mínimo
    if minimo is not None:

        mask = valores.notna() & (valores < minimo)

        for indice in df.index[mask]:
            errores[indice].append(
                f"{columna}: '{df.at[indice, columna]}' es menor que {minimo}."
            )

    # Mayor que el máximo
    if maximo is not None:

        mask = valores.notna() & (valores > maximo)

        for indice in df.index[mask]:
            errores[indice].append(
                f"{columna}: '{df.at[indice, columna]}' es mayor que {maximo}."
            )
```

Hacemos lo mismo para los floats.

```python
# Floats.
for columna, (minimo, maximo) in FLOATS.items():

    valores = pd.to_numeric(df[columna], errors="coerce")

    mask = valores.isna()

    for indice in df.index[mask]:
        errores[indice].append(
            f"{columna}: '{df.at[indice, columna]}' no es un número decimal."
        )

    if minimo is not None:

        mask = valores.notna() & (valores < minimo)

        for indice in df.index[mask]:
            errores[indice].append(
                f"{columna}: '{df.at[indice, columna]}' es menor que {minimo}."
            )

    if maximo is not None:

        mask = valores.notna() & (valores > maximo)

        for indice in df.index[mask]:
            errores[indice].append(
                f"{columna}: '{df.at[indice, columna]}' es mayor que {maximo}."
            )
```

Hacemos lo mismo para los strings.

```python
# Strings.
for columna, permitidos in STRINGS.items():

    # No es una cadena
    mask = ~df[columna].map(lambda x: isinstance(x, str))

    for indice in df.index[mask]:
        errores[indice].append(
            f"{columna}: '{df.at[indice, columna]}' no es una cadena de texto."
        )

    # Valor no permitido
    mask = (
        df[columna].map(lambda x: isinstance(x, str))
        & ~df[columna].isin(permitidos)
    )

    for indice in df.index[mask]:
        errores[indice].append(
            f"{columna}: '{df.at[indice, columna]}' "
            f"no pertenece a {sorted(permitidos)}."
        )
```

Hacemos lo mismo para los booleanos.

```python
#Booleanos.
for columna in BOOLEANOS:
    mask = ~df[columna].map(lambda x: isinstance(x, bool))
    for indice in df.index[mask]:
        errores[indice].append(
            f"{columna}: '{df.at[indice, columna]}' no es un valor booleano."
        )
```

Hacemos lo mismo para los DNIs. Comprobamos que tengan 8 numeros y una letra.

```python
#DNI.
dni = (
    df["DNI"]
    .astype("string")
    .str.strip()
    .str.upper()
)

mask = ~dni.str.fullmatch(r"\d{8}[A-Z]", na=False)

for indice in df.index[mask]:
    errores[indice].append(
        f"DNI: '{df.at[indice, 'DNI']}' no tiene el formato correcto."
    )
```

Comprobamos que las columnas Fills y No Fills no sean iguales, pues sería una contradicción.

```python
#Fills y no fills.
mask = (
    df["Fills"].notna()
    & df["No Fills"].notna()
    & (df["Fills"] == df["No Fills"])
)

for indice in df.index[mask]:

    errores[indice].append(
        f"Fills={df.at[indice,'Fills']} y "
        f"No Fills={df.at[indice,'No Fills']} "
        "deberían ser valores opuestos."
    )
```

Mostramos los errores encontrados en cada fila del DataFrame.

```python
for fila, lista in errores.items():

    print(f"\nFila {fila}")

    for error in lista:
        print(f"  • {error}")
```
