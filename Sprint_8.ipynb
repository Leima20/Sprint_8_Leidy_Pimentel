Definimos una función auxiliar para reparar los errores de codificación.

Al inspeccionar el dataset se detectan errores de codificación tanto en los nombres de algunas columnas como en varios valores de texto. Algunos ejemplos son `CÃ rrec` en lugar de `Càrrec`, `AdriÃ ` en lugar de `Adrià` o `MuÃ±oz` en lugar de `Muñoz`.

Para corregir estos problemas, se define una función que intenta reparar los textos mediante una conversión entre las codificaciones CP1252 y UTF-8. Si la conversión no es posible, se devuelve el valor original para evitar modificar información que ya sea correcta.

```python
def reparar_codificacion(texto) -> str:
    """
    Corrige posibles errores de codificación en una cadena de texto.

    Intenta reconstruir caracteres mal codificados convirtiendo desde
    Windows-1252 (cp1252) a UTF-8. Si la conversión no es posible,
    devuelve el texto original.

    Args:
        texto (str): Texto a reparar.

    Returns:
        str: Texto reparado o el original si no requiere corrección.
    """

    try:
        return texto.encode("cp1252").decode("utf-8")

    except (UnicodeEncodeError, UnicodeDecodeError):
        return texto
```

 **Corrección de la codificación en las cabeceras**

Una vez definida la función de reparación, se aplica a los nombres de las columnas del dataset.

Posteriormente, se comparan las cabeceras originales con las reparadas para identificar cuáles presentaban problemas de codificación. Si se detectan diferencias, se actualizan los nombres de las columnas utilizando la versión corregida.

De esta forma se garantiza que los nombres de las variables sean legibles y puedan utilizarse correctamente durante el resto del análisis.

```python
# Guardar las cabeceras originales
cabeceras_originales = df.columns

print(f"Cabeceras originales: {cabeceras_originales.tolist()}")
```

```python
# Reparar la codificación de las cabeceras
cabeceras_reparadas = cabeceras_originales.map(
    lambda columna: (
        reparar_codificacion(columna)
        if isinstance(columna, str)
        else columna
    )
)
```

```python
# Identificar las cabeceras que han sido modificadas
cabeceras_modificadas = cabeceras_originales[
    cabeceras_originales != cabeceras_reparadas
]
```

```python
# Sustituir las cabeceras si se han encontrado errores
if len(cabeceras_modificadas) > 0:

    print(
        "Se han reparado las siguientes cabeceras debido "
        "a problemas de codificación:"
    )

    print(cabeceras_modificadas.tolist())

    df.columns = cabeceras_reparadas

else:

    print(
        "No se han detectado problemas de codificación "
        "en las cabeceras."
    )
```

**Corrección de la codificación en los valores de texto**

Una vez corregidas las cabeceras, se aplica el mismo procedimiento a todas las columnas de tipo texto del dataset.

Para cada columna se comparan los valores originales con los valores reparados. Si se detectan diferencias, únicamente se sustituyen aquellos registros afectados por problemas de codificación, manteniendo el resto de la información sin modificaciones.

Además, se informa del número de valores corregidos en cada columna para documentar las transformaciones realizadas durante el proceso de limpieza de datos.

```python
# Revisar todas las columnas de tipo texto
for columna in df.select_dtypes(include=["object", "string"]):

    # Guardar la serie original
    serie_original = df[columna]

    # Aplicar la reparación únicamente a los valores de texto
    serie_reparada = serie_original.map(
        lambda valor: (
            reparar_codificacion(valor)
            if isinstance(valor, str)
            else valor
        )
    )

    # Contar cuántos valores han cambiado
    valores_reparados = (serie_original != serie_reparada).sum()

    # Sustituir la columna únicamente si se han realizado cambios
    if valores_reparados > 0:

        print(
            f"Se han reparado {valores_reparados} valores "
            f"en la columna '{columna}'."
        )

        df[columna] = serie_reparada
```

## Validación de la unicidad del DNI

El DNI se considera el identificador único de cada empleado, por lo que no deberían existir dos registros con el mismo valor.

Para comprobar su unicidad, se identifican todos los DNIs repetidos y se revisa la información asociada a cada uno de ellos.

En este dataset, los registros que comparten un mismo DNI presentan información inconsistente y no es posible determinar cuál de ellos es el correcto. Ante la imposibilidad de decidir de forma objetiva qué registro conservar, se opta por eliminar todos los registros asociados a los DNIs duplicados.

Esta decisión evita mantener información potencialmente errónea que pueda afectar a los análisis posteriores.

```python
# Identificar todos los registros con un DNI duplicado
duplicados = df[df.duplicated(subset=["DNI"], keep=False)]

if duplicados.empty:

    print("No se han encontrado DNIs duplicados.")

else:

    for dni, grupo in duplicados.groupby("DNI"):

        print(f"\nRegistros con el DNI duplicado {dni}:")

        display(grupo)

        # Eliminar todos los registros asociados a ese DNI
        df.drop(index=grupo.index, inplace=True)

        print(
            f"Se han eliminado {len(grupo)} registros "
            f"con el DNI {dni}."
        )
```

Antes de realizar las validaciones, se definen las reglas que debe cumplir cada tipo de dato. Estas reglas establecen los rangos permitidos para las variables numéricas, las categorías válidas para las variables de texto y las columnas que deben contener valores booleanos. Posteriormente, estas definiciones se utilizarán para comprobar la calidad de los datos del dataset.

```python
# Reglas de validación

ENTEROS = {
    "Dia de Naixement": (1, 31),
    "Mes de Naixement": (1, 12),
    "Any de Naixement": (1900, 2026),
    "Salari mensual": (0, None),
    "Nombre_fills": (0, None),
    "Km_anuals": (0, None),
}

FLOATS = {
    "Consum_mitja_L_100km": (0, None),
    "Temperatura_mitjana_ciutat": (-273.15, None),
}

STRINGS = {
    "Gènere": {"H", "D", "Home", "Dona"},
    "Grup Professional": {
        "Grup A",
        "Grup B",
        "Grup C",
        "Grup D",
    },
}

BOOLEANOS = [
    "Te_cotxe",
    "Fills",
    "No Fills",
]
```

Valoracion de tipos de datos y rangos permitidos

En este bloque se valida que los datos almacenados en cada columna cumplan con el formato esperado. Para ello se definen reglas según el tipo de variable:

Variables enteras (ENTEROS): se comprueba que los valores sean numéricos y se encuentren dentro de un rango válido.
Variables decimales (FLOATS): se verifica que sean números decimales y respeten los límites establecidos.
Variables categóricas (STRINGS): se comprueba que el valor pertenezca al conjunto de categorías permitidas.

Cada incidencia detectada se almacena en un diccionario (errores), permitiendo identificar todos los problemas de cada registro antes de decidir qué acciones tomar sobre ellos.

```python
errores = defaultdict(list)

# Validación de variables enteras

for columna, (minimo, maximo) in ENTEROS.items():

    valores = pd.to_numeric(df[columna], errors="coerce")

    # No es un número entero
    mask = valores.isna()

    for indice in df.index[mask]:
        errores[indice].append(
            f"{columna}: '{df.at[indice, columna]}' no es un número entero."
        )

    # Menor que el mínimo
    if minimo is not None:

        mask = valores.notna() & (valores < minimo)

        for indice in df.index[mask]:
            errores[indice].append(
                f"{columna}: '{df.at[indice, columna]}' es menor que {minimo}."
            )

    # Mayor que el máximo
    if maximo is not None:

        mask = valores.notna() & (valores > maximo)

        for indice in df.index[mask]:
            errores[indice].append(
                f"{columna}: '{df.at[indice, columna]}' es mayor que {maximo}."
            )
```

```python
# Floats.
for columna, (minimo, maximo) in FLOATS.items():

    valores = pd.to_numeric(df[columna], errors="coerce")

    mask = valores.isna()

    for indice in df.index[mask]:
        errores[indice].append(
            f"{columna}: '{df.at[indice, columna]}' no es un número decimal."
        )

    if minimo is not None:

        mask = valores.notna() & (valores < minimo)

        for indice in df.index[mask]:
            errores[indice].append(
                f"{columna}: '{df.at[indice, columna]}' es menor que {minimo}."
            )

    if maximo is not None:

        mask = valores.notna() & (valores > maximo)

        for indice in df.index[mask]:
            errores[indice].append(
                f"{columna}: '{df.at[indice, columna]}' es mayor que {maximo}."
            )
```

```python
# Strings.
for columna, permitidos in STRINGS.items():

    # No es una cadena
    mask = ~df[columna].map(lambda x: isinstance(x, str))

    for indice in df.index[mask]:
        errores[indice].append(
            f"{columna}: '{df.at[indice, columna]}' no es una cadena de texto."
        )

    # Valor no permitido
    mask = (
        df[columna].map(lambda x: isinstance(x, str))
        & ~df[columna].isin(permitidos)
    )

    for indice in df.index[mask]:
        errores[indice].append(
            f"{columna}: '{df.at[indice, columna]}' "
            f"no pertenece a {sorted(permitidos)}."
        )
```

```python
#Booleanos.
for columna in BOOLEANOS:
    mask = ~df[columna].map(lambda x: isinstance(x, bool))
    for indice in df.index[mask]:
        errores[indice].append(
            f"{columna}: '{df.at[indice, columna]}' no es un valor booleano."
        )
```

```python
#DNI.
dni = (
    df["DNI"]
    .astype("string")
    .str.strip()
    .str.upper()
)

mask = ~dni.str.fullmatch(r"\d{8}[A-Z]", na=False)

for indice in df.index[mask]:
    errores[indice].append(
        f"DNI: '{df.at[indice, 'DNI']}' no tiene el formato correcto."
    )
```

```python
#Fills y no fills.
mask = (
    df["Fills"].notna()
    & df["No Fills"].notna()
    & (df["Fills"] == df["No Fills"])
)

for indice in df.index[mask]:

    errores[indice].append(
        f"Fills={df.at[indice,'Fills']} y "
        f"No Fills={df.at[indice,'No Fills']} "
        "deberían ser valores opuestos."
    )
```

```python
print(f"Se han detectado errores en {len(errores)} registros.")
for fila, lista in errores.items():

    print(f"\nFila {fila}")

    for error in lista:
        print(f"  • {error}")
```

Comprobar los tipos de datos dentro de las variables categóricas

Aunque una columna tenga tipo object, esto no garantiza que todos sus valores sean cadenas de texto. Se comprueba el tipo de los datos almacenados en cada variable para identificar posibles mezclas de tipos (por ejemplo, cadenas, booleanos o valores nulos) y aplicar posteriormente las transformaciones únicamente cuando corresponda.

```python
columnas_categoricas = df.select_dtypes(include="object").columns

for col in columnas_categoricas:
    print(f"\n--- {col} ---")
    print(df[col].apply(type).value_counts())
```

Se eliminan los espacios al inicio y al final de las cadenas de texto y se convierten todos los valores a minúsculas. Esta transformación evita que diferencias de formato generen categorías distintas que representen el mismo concepto. La comprobación previa de los tipos de datos permite aplicar la transformación únicamente a los valores de tipo str, preservando el resto de tipos de datos.

```python
df[columnas_categoricas] = df[columnas_categoricas].apply(
    lambda col: col.map(
        lambda x: x.strip().lower() if isinstance(x, str) else x
    )
)
```
